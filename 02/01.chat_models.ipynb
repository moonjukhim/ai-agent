You can use the same interface to work with different LLMs.

```python
from langchain_openai import OpenAI
#from langchain_google_genai import GoogleGenerativeAI

# initialize Gemini
# gemini_pro = GoogleGenerativeAI(model="gemini-2.0-flash")

# or: initialize OpenAI model
openai_llm = OpenAI()

response = openai_llm.invoke("오늘 날씨는 어때?")
print(response)
```

#### FakeListLLM을 사용하여 API 호출 없이 테스트

```python
from langchain_core.language_models.fake import FakeListLLM

fake_llm = FakeListLLM(
    responses=["Hello"]
)

print(fake_llm.invoke("Anything"))
```

The default interface to work with LLMs if the Chat interface though.

```python
# from langchain_anthropic import ChatAnthropic
from langchain_openai.chat_models import ChatOpenAI

from langchain_core.messages import SystemMessage, HumanMessage

# specifying a model:
chat = ChatOpenAI(model='gpt-4o-mini')
# chat = ChatAnthropic(model="claude-3-opus-20240229")

# messages:
messages = [
    SystemMessage(content="당신은 유능한 프로그래머입니다."),
    HumanMessage(content="Hello, World를 출력하는 코드를 작성해줘.")
]
response = chat.invoke(messages)
```

```python
print(response.content)
```
